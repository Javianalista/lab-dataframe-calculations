## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
#sub1
unique_customer_amount = orders.groupby('CustomerID').sum('amount_spend')
unique_customer_amount['CustomerID'] = unique_customer_amount.index
display(unique_customer_amount)
```

```python
#sub2
percentile_95 = np.percentile(unique_customer_amount['amount_spent'], 95,interpolation='linear')
display(percentile_95)
percentile_75 = np.percentile(unique_customer_amount['amount_spent'], 75,interpolation='linear')
display(percentile_75)
```

```python
vip_customers = unique_customer_amount[unique_customer_amount['amount_spent'] >= percentile_95]
preferred_customers = unique_customer_amount[(unique_customer_amount['amount_spent'] < percentile_95) & (unique_customer_amount['amount_spent'] >= percentile_75)]
list_vip = vip_customers['CustomerID'].tolist()
list_preferred = preferred_customers['CustomerID'].tolist()
```

```python
vips = orders[orders['CustomerID'].isin(list_vip)]
vips
```

```python
preferred = orders[orders['CustomerID'].isin(list_preferred)]
preferred
```

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
vip_c = vips[['CustomerID','Country']]
vip_c = vip_c.reset_index(drop =True)
vip_c
```

```python
vip_c['CustomerID'].nunique()
```

```python
position = [] # Here i'm saving all the positions i need to obtain all the unique customers
unique_cust = [] # Here i'm saving all the unique customers to use it into the for loop
for i in range(len(vip_c['CustomerID'])):
    if vip_c['CustomerID'][i] not in unique_cust:
        position = position + [i]
        unique_cust = unique_cust + [vip_c['CustomerID'][i]]
len(position)
```

```python
unique_vips = vip_c[vip_c.index.isin(position)]
unique_vips
```

```python
unique_vips['Country'].value_counts()  # Here is the list 
```

```python
# Another way
unique_vips1 = unique_vips.groupby(['Country']).count()
unique_vips1 = unique_vips1.sort_values('CustomerID',ascending=False)
unique_vips1
```
